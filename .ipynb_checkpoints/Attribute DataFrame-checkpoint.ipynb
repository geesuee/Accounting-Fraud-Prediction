```python
import pandas as pd
```

```python
def df_attribute(attribute_name):
    attribute = pd.read_excel('data/attri/{}.xls'.format(attribute_name) ,skiprows = [0,1,2,3,4])
    attribute.set_index('Unnamed: 2', inplace=True)
    attribute = attribute[['2014/Annual', '2015/Annual', '2016/Annual', '2017/Annual', '2018/Annual','2019/Annual']]
    attribute.columns=['2014','2015','2016','2017','2018','2019']
    attribute = pd.DataFrame(attribute.drop('Name').stack(dropna=False))
    attribute.columns=[attribute_name]
    attribute.reset_index(inplace=True)
    attribute.rename(columns = {'Unnamed: 2' : '기업명'}, inplace = True)
    attribute.rename(columns = {'level_1' : '년도'}, inplace = True)
    return attribute

def df_attribute1(attribute_name):
    attribute = pd.read_excel('data/attri/{}.xls'.format(attribute_name) ,skiprows = [0,1,2,3,4])
    attribute.set_index('Unnamed: 2', inplace=True)
    attribute = attribute[['2014/Annual', '2015/Annual', '2016/Annual', '2017/Annual', '2018/Annual','2019/Annual']]
    attribute.columns=['2014','2015','2016','2017','2018','2019']
    attribute = pd.DataFrame(attribute.drop('Name').stack(dropna=False))
    attribute.columns=[attribute_name]
    attribute.reset_index(inplace=True)
    return attribute

def df_attribute2(attribute_name):
    attribute = pd.read_excel('data/attri/{}.xls'.format(attribute_name) ,skiprows = [0,1,2,3,4])
    attribute.set_index('Unnamed: 2', inplace=True)
    attribute = attribute[['2014', '2015', '2016', '2017', '2018','2019']]

    attribute = pd.DataFrame(attribute.drop('Name').stack(dropna=False))
    attribute.columns=[attribute_name]
    attribute.reset_index(inplace=True)
    return attribute
```

```python
매출액 = df_attribute('매출액')
당기순이익 = df_attribute1('당기순이익')
매출액영업이익율 = df_attribute1('매출액영업이익율')
매출채권1 = df_attribute1('매출채권1')
매출채권2 = df_attribute1('매출채권2')
매출채권회전율 = df_attribute1('매출채권회전율')
부채비율 = df_attribute1('부채비율')
비금융계_재고자산 = df_attribute1('비금융계_재고자산')
영업활동으로인한현금흐름 = df_attribute1('영업활동으로인한현금흐름')
유동부채 = df_attribute1('유동부채')
유동비율 = df_attribute1('유동비율')
유동자산 = df_attribute1('유동자산')
자기자본순이익율 = df_attribute1('자기자본순이익율')
자산총계 = df_attribute1('자산총계')
재고자산회전율1 = df_attribute1('재고자산회전율1')
재고자산회전율2 = df_attribute1('재고자산회전율2')
종업원수 = df_attribute1('종업원수')
종업원수증가율 = df_attribute1('종업원수증가율')

감사법인 = df_attribute2('감사법인')
감사의견 = df_attribute2('감사의견')
대주주명 = df_attribute2('대주주명')
KIS신용평점 = df_attribute2('KIS신용평점')
```

```python
all= pd.concat([매출액, 당기순이익, 매출액영업이익율, 매출채권1, 매출채권2, 매출채권회전율, 부채비율, 비금융계_재고자산,
                영업활동으로인한현금흐름, 유동부채, 유동비율, 유동자산, 자기자본순이익율, 자산총계, 재고자산회전율1,
                재고자산회전율2, 종업원수, 종업원수증가율, 감사법인, 감사의견, 대주주명, KIS신용평점], axis = 1)
all
```

```python
all.drop(['level_1'], axis = 1, inplace = True)
all.drop(['Unnamed: 2'], axis = 1, inplace = True)
all
```
