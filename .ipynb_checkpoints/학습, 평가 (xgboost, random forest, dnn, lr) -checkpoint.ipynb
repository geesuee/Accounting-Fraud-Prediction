# 데이터를 train set 과 test set으로 나누기
## - over sampling을 통해 불균형 데이터 문제 해결

```python
# Use a utility from sklearn to split and shuffle our dataset.

y = cleaned_df.pop('AO')
X = cleaned_df

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0,stratify=y) 
```

```python
X_train.columns
```

```python
from imblearn.over_sampling import SMOTENC
cf = [13,15,16,17,18,19]
smotenc = SMOTENC(categorical_features=cf)
X_train_over,y_train_over = smotenc.fit_sample(X_train,y_train)
```

## 정확도, 재현율, auc 함수

```python
from sklearn.metrics import accuracy_score, recall_score, roc_auc_score
```

```python
def get_clf_eval(y_test, pred):
    accuracy = accuracy_score(y_test, pred)
    recall = recall_score(y_test, pred)
    auc = roc_auc_score(y_test, pred)
    print('정확도: {:.2f}'.format(accuracy))
    print('재현율: {:.2f}'.format(recall))
    print('auc: {:.2f}'.format(auc))
```

## Random Forest

```python
neg, pos = np.bincount(y)
total = neg + pos
print('Examples:\n    Total: {}\n    Positive: {} ({:.2f}% of total)\n'.format(
    total, pos, 100 * pos / total))


weight_for_0 = (1 / neg)*(total)/2.0 
weight_for_1 = (1 / pos)*(total)/2.0

class_weights = {0: weight_for_0, 1: weight_for_1}
```

```python
# Random Forest w/ giving weights to classes

rf_clf = RandomForestClassifier(n_estimators = 500, max_depth = 7, class_weight=class_weights)
rf_clf.fit(X_train, y_train)
recall_rf = recall_score(y_test, rf_clf.predict(X_test))
get_clf_eval(y_test, rf_clf.predict(X_test))
```

```python
# Random Forest with oversampling(SmoteNC)

rf_clf2 = RandomForestClassifier(n_estimators = 500, max_depth = 7)
rf_clf2.fit(X_train_over, y_train_over)
recall_rf2 = recall_score(y_test, rf_clf2.predict(X_test))
get_clf_eval(y_test, rf_clf2.predict(X_test))
```

## xgboost

```python
xgb_clf = XGBClassifier(booster='dart', rate_drop=0.2, alpha=0.5, eval_metric='logloss', n_estimators=200, 
                        max_depth=4, eta=0.2, colsample_bytree=0.9, subsample=0.8, max_delta_step = 1)
xgb_clf.fit(X_train_over, y_train_over)
recall_xgb = recall_score(y_test, xgb_clf.predict(X_test))
get_clf_eval(y_test, xgb_clf.predict(X_test))
```

```python
def plot_tree(xgb_model, filename):
    
    import xgboost as xgb
    import os
    
    gvz = xgb.to_graphviz(xgb_model)
    _, file_extension = os.path.splitext(filename)
    format = file_extension.strip('.').lower()
    data = gvz.pipe(format=format)
    full_filename = filename
    with open(full_filename, 'wb') as f:
        f.write(data)
```

## DNN

```python
Y = np.array(y)
Y = to_categorical(Y)

X = cleaned_df

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=0,stratify=y) 

# Form np arrays of labels and features.

y_train_labels = np.array(y_train)
bool_train_labels = y_train[:,[1]].reshape(-1) != 0
y_test_labels = np.array(y_test)


X_train_features = np.array(X_train)
X_test_features = np.array(X_test)
```

```python
def make_model():
    model = Sequential()
    model.add(Dense(units=40, input_dim=20, activation='relu'))
    model.add(Dense(units=38, activation='relu'))
    model.add(Dense(units=30, activation='relu'))
    model.add(Dense(units=25, activation='relu'))
    model.add(Dense(units=20, activation='relu'))
    model.add(Dense(units=18, activation='relu'))
    model.add(Dense(units=15, activation='relu'))
    model.add(Dense(units=7, activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(units=2, activation='softmax'))

    model.compile(loss=keras.losses.BinaryCrossentropy(), optimizer="adam", metrics = ['accuracy'])

    return model
```

```python
EPOCHS = 100
BATCH_SIZE = 2048

early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_auc', 
    verbose=1,
    patience=10,
    mode='max',
    restore_best_weights=True) 
```

```python
model = make_model()
model.summary()
```

```python
baseline_history = model.fit(X_train_features, y_train_labels, batch_size = BATCH_SIZE, epochs = EPOCHS, class_weight=class_weights)
```

```python
model.evaluate(X_test_features, y_test_labels)
```

```python
# Keras 내장 metric 대신 기존 evaluation 함수 활용

recall_dnn = recall_score(y_test_labels.T[1], np.argmax(model.predict(X_test_features), 1))
get_clf_eval(y_test_labels.T[1], np.argmax(model.predict(X_test_features), 1))
```

```python
pd.DataFrame([recall_rf, recall_xgb, recall_lr, recall_dnn], index = ['Random Forest', 'XGBoost', 'Logistic Regression', 'DNN'], columns = ['Recall Score']).sort_values(by = 'Recall Score', ascending = False).plot(kind='barh', xlim = (0.6, 0.9), )
```

```python
pd.DataFrame([recall_rf, recall_xgb, recall_lr, recall_dnn], index = ['Random Forest', 'XGBoost', 'Logistic Regression', 'DNN'], columns = ['Recall Score']).sort_values(by = 'Recall Score', ascending = False)
```

```python
df_models = pd.DataFrame([recall_rf, recall_xgb, recall_lr, recall_dnn], index = ['Random Forest', 'XGBoost', 'Logistic Regression', 'DNN'], columns = ['Recall Score']).sort_values(by = 'Recall Score', ascending = False)

ax = df_models.plot(kind = 'barh', color=[['gold', 'gray', 'gray', 'gray']], title = 'Recall Score', legend = False, xlim = (0.6, 0.9))
plt.show()
```

```python
주식수익률.년도 = 주식수익률.년도.astype(int)
주식수익률.주식수익률 = 주식수익률.주식수익률.astype(float)
```

```python
df_수익률 = pd.merge(주식수익률, df_prob, how = 'inner').sort_values(by = 'probability')
```

```python
df_수익률.head()
```

```python
df_수익률 = pd.concat([df_수익률, pd.qcut(df_수익률.probability, q=75)], axis = 1)
df_수익률.columns = ['기업명', '년도', '주식수익률', 'probability', 'class', 'bin']
```

```python
Q1 = df_수익률.주식수익률.quantile(0.25)
Q3 = df_수익률.주식수익률.quantile(0.75)
IQR = Q3 - Q1

df_수익률 = df_수익률[~((df_수익률.주식수익률 < (Q1 - 3 * IQR)) | (df_수익률.주식수익률 > (Q3 + 3 * IQR)))]
```

```python
fig = plt.figure()
df_수익률.groupby('bin').주식수익률.mean().plot(color = 'red')
plt.xlabel('비적정일 확률')
plt.ylabel('주식수익률')
plt.show()
```

```python
fig = plt.figure()
df_수익률.groupby('class').주식수익률.mean().plot(kind = 'bar', color = ['navy', 'crimson'])
plt.xlabel('비적정여부')
plt.ylabel('주식수익률')
plt.xticks((0, 1))
plt.show()

# df_수익률.groupby('bin').주식수익률.mean().plot()
```
