```python
def df_attribute(attribute_name):
    attribute = pd.read_excel('data/attri/{}.xls'.format(attribute_name) ,skiprows = [0,1,2,3,4])
    attribute['Unnamed: 2'] = attribute['Unnamed: 2'] + attribute['Unnamed: 0']
    attribute.set_index('Unnamed: 2', inplace=True)
    attribute = attribute[['2009/Annual', '2010/Annual', '2011/Annual', '2012/Annual', '2013/Annual','2014/Annual', '2015/Annual', '2016/Annual', '2017/Annual', '2018/Annual','2019/Annual']]
    attribute.columns=['2009', '2010', '2011', '2012', '2013', '2014','2015','2016','2017','2018','2019']
    attribute = pd.DataFrame(attribute.drop('NameKIS').stack(dropna=False))
    attribute.columns=[attribute_name]
    attribute.reset_index(inplace=True)
    attribute.rename(columns = {'Unnamed: 2' : '기업명'}, inplace = True)
    attribute.rename(columns = {'level_1' : '년도'}, inplace = True)
    return attribute


# year/Annual 형식 (kis code 포함)
def df_attribute1(attribute_name):
    attribute = pd.read_excel('data/attri/{}.xls'.format(attribute_name) ,skiprows = [0,1,2,3,4])
    attribute['Unnamed: 2'] = attribute['Unnamed: 2'] + attribute['Unnamed: 0']
    attribute.set_index('Unnamed: 2', inplace=True)
    attribute = attribute[['2009/Annual', '2010/Annual', '2011/Annual', '2012/Annual', '2013/Annual','2014/Annual', '2015/Annual', '2016/Annual', '2017/Annual', '2018/Annual','2019/Annual']]
    attribute.columns=['2009', '2010', '2011', '2012', '2013', '2014','2015','2016','2017','2018','2019']
    attribute = pd.DataFrame(attribute.drop('NameKIS').stack(dropna=False))
    attribute.columns=[attribute_name]
    attribute.reset_index(inplace=True)

    return attribute

# year 형식 (kis code 포함)
def df_attribute2(attribute_name):
    attribute = pd.read_excel('data/attri/{}.xls'.format(attribute_name) ,skiprows = [0,1,2,3,4])
    attribute['Unnamed: 2'] = attribute['Unnamed: 2'] + attribute['Unnamed: 0']
    attribute.set_index('Unnamed: 2', inplace=True)
    attribute = attribute[['2009', '2010', '2011', '2012', '2013', '2014','2015','2016','2017','2018','2019']]
    attribute = pd.DataFrame(attribute.drop('NameKIS').stack(dropna=False))
    attribute.columns=[attribute_name]
    attribute.reset_index(inplace=True)

    return attribute

# 2015-2018 year 형식
def df_attribute3(attribute_name):
    attribute = pd.read_excel('data/attri/{}.xls'.format(attribute_name) ,skiprows = [0,1,2,3,4])
    attribute['Unnamed: 2'] = attribute['Unnamed: 2'] + attribute['Unnamed: 0']
    attribute.set_index('Unnamed: 2', inplace=True)
    attribute = attribute[['2009', '2010', '2011', '2012', '2013', '2014','2015','2016','2017','2018','2019']]
    attribute = pd.DataFrame(attribute.drop('NameKIS').stack(dropna=False))
    attribute.columns=[attribute_name]
    attribute.reset_index(inplace=True)

    return attribute
```

## (당기-전기)/전기 함수

```python
def fluctuation_rate1(df):
    변동_at = df.copy()
    변동_at = 변동_at.loc[:, '2009':'2019']
    for idx in range(len(df.columns)-1):
        변동_at[df.columns[idx+1]] =\
        (df[df.columns[idx+1]]-df[df.columns[idx]])/df[df.columns[idx]]
    return 변동_at        
```

## 당기/전기 함수

```python
def fluctuation_rate2(df):
    변동_at = df.copy()
    변동_at = 변동_at.loc[:, '2009':'2019']
    for idx in range(len(df.columns)-1):
        변동_at[df.columns[idx+1]] =\
        df[df.columns[idx+1]]/df[df.columns[idx]]
    return 변동_at       
```

## 당기-전기 함수

```python
def fluctuation_rate3(df):
    변동_at = df.copy()
    변동_at = 변동_at.loc[:, '2009':'2019']
    for idx in range(len(df.columns)-1):
        변동_at[df.columns[idx+1]] =\
        df[df.columns[idx+1]]-df[df.columns[idx]]
    return 변동_at    
```
