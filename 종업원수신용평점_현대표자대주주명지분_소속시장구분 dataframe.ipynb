```python
소속시장구분 = 소현신['Unnamed: 3']
소속시장구분 = pd.DataFrame(소속시장구분.drop('Name'))
소속시장구분.columns = ['소속시장구분']
소속시장구분.reset_index(inplace=True)
소속시장구분
```

```python
현대표자 = 소현신['Unnamed: 10']
현대표자 = pd.DataFrame(현대표자.drop('Name'))
현대표자.columns = ['현대표자']
현대표자.reset_index(inplace=True)
현대표자
```

```python
대주주명지분['level_1'].value_counts()
```

```python
대주주명지분 = 소현신[['2015', '2016', '2017', '2018', '2019']]
대주주명지분 = pd.DataFrame(대주주명지분.drop('Name').stack(dropna=False))
대주주명지분.columns = ['대주주명지분']
대주주명지분.reset_index(inplace=True)
대주주명지분['level_1'] = 대주주명지분['level_1'].map({
    '2015': '2015/Annual',
    '2016': '2016/Annual',
    '2017': '2017/Annual',
    '2018': '2018/Annual',
    '2019': '2019/Annual',
})
대주주명지분
```

```python
종업원수['level_1'].value_counts()
```

```python
종업원수 = 소현신[['2014/Annual', '2015/Annual', '2016/Annual', 
                   '2017/Annual', '2018/Annual', '2019/Annual']]
종업원수 = pd.DataFrame(종업원수.drop('Name').stack(dropna=False))
종업원수.columns = ['종업원수']
종업원수.reset_index(inplace=True)
종업원수
```

```python
신용평점['level_1'].value_counts()
```

```python
신용평점 = 소현신[['2015.1', '2016.1', '2017.1', '2018.1', '2019.1']]
신용평점 = pd.DataFrame(신용평점.drop('Name').stack(dropna=False))
신용평점.columns = ['신용평점']
신용평점.reset_index(inplace=True)
신용평점['level_1'] = 신용평점['level_1'].map({
    '2015.1': '2015/Annual',
    '2016.1': '2016/Annual',
    '2017.1': '2017/Annual',
    '2018.1': '2018/Annual',
    '2019.1': '2019/Annual',
})  # 2014년도가 없음
신용평점
```

```python
종신merge = pd.merge(종업원수, 신용평점, how='left')
종신merge
# 2014년 종업원수쪽에 신용평점 NaN
```
